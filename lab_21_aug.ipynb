```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# unsupervised
# Load CSV
df = pd.read_csv("datasets/Mall_Customers.csv")

# Select features for clustering
X = df[["Annual Income (k$)", "Spending_Score"]]

# Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=0)
df["Cluster"] = kmeans.fit_predict(X)

# Plot results
plt.scatter(X["Annual Income (k$)"], X["Spending_Score"], 
            c=df["Cluster"], cmap="viridis")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c="red", marker="X", label="Centroids")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
#supervised
# Regression, Classifcation means asigning groups from Training data type.
# Features
X = df[["Age", "Annual Income (k$)"]]
y = df["Spending_Score"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply KNN Regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict
y_pred = knn.predict(X_test)

# Metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("MAE:", mae)
print("R² Score:", r2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Encode Gender as numeric (Male=0, Female=1)
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})

# print(df["Gender"].isna().sum())  # check for leftover NaNs
# Features and target
X = df[["Gender", "Age", "Annual Income (k$)"]]
y = df["Spending_Score"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply KNN Regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict
y_pred = knn.predict(X_test)

# Metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("MAE:", mae)
print("R² Score:", r2)
```
